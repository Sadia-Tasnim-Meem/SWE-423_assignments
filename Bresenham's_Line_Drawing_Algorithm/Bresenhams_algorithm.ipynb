```python
import matplotlib.pyplot as plt
```

```python
def bresenham(x0,y0,xf,yf):
    x_co = [x0,xf]
    y_co = [y0,yf]
    plt.plot(x_co, y_co,linestyle='-', color = 'red')
    dx = xf-x0
    dy = yf-y0
    
    if abs(dy) > 0 and abs(dy) < abs(dx):
        x = x0;
        if dy < 0:
            y = -y0
        else:
            y = y0
        
        
        dT = 2*(dy - dx); #taking the pixel on the upper right corner
        dS = 2*dy; #taking the pixel on the right
        d = 2*dy - dx; 
        
        x_values = [x]
        y_values = [y]
        
        points = [(x,y)]
        
        while x < xf:
            x += 1
            if d < 0:
                d += dS
            else:
                y += 1
                d += dT
            
            x_values.append(x)
            y_values.append(y)
            points.append((x,y))
    elif abs(dy) > abs(dx):
        x = x0
        y = y0
        dT = 2*(dx-dy)
        dS = 2*dx
        d = 2*dx - dy
        
        x_values = [x]
        y_values = [y]
        points = [(x,y)]
        
        while y < yf:
            y += 1
            if d < 0:
                d += dS
            
            else:
                x += 1
                d += dT
            
            x_values.append(x)
            y_values.append(y)
            points.append((x,y))
    
    print("The points are as follows:")
    print(points)
    
    plt.title("Bresenham's Algorithm")
    plt.xlabel("X Axis")
    plt.ylabel("Y Axis")
    plt.scatter(x_values, y_values)
    plt.plot(x_values, y_values)
    plt.axis("square")
    plt.grid()
    plt.show()
        
```

## Case 1: initial point(1,1) and final point(8,4)

```python
bresenham(1,1,8,4)
```

## Case 2: initial point: (1,1) and final point(4,8)

```python
bresenham(1,1,4,8)
```
